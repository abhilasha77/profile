# Question and answer
**Q1:How does the revenue generated from document registration vary across districts in Telangana? List down the top 5 districts that showed the highest document registration revenue growth between FY 2019 and 2022.**

```python
import matplotlib.pyplot as mtplot
import seaborn as sb
```

```python
# Filter data for FY 2019 and FY 2022
Data_2019 = Data[Data['month'].between('2019-04-01', '2020-03-31')]
Data_2022 = Data[Data['month'].between('2022-04-01', '2023-03-31')]

# Group by district and calculate total revenue for FY 2019 and FY 2022
revenue_2019 = Data_2019.groupby('district')['documents_registered_rev'].sum()
revenue_2022 = Data_2022.groupby('district')['documents_registered_rev'].sum()

# Calculate revenue growth
revenue_growth = (revenue_2022 - revenue_2019) / revenue_2019 * 100

# Sort the districts by revenue growth in descending order and get the top 5
top_5_districts = revenue_growth.sort_values(ascending=False).head(5)

# Print the top 5 districts with the highest revenue growth
print("Top 5 Districts with Highest Document Registration Revenue Growth (FY 2019 to FY 2022):")
print(top_5_districts)









```

**Bar chart to understand better vision**

```python
mtplot.figure(figsize=(4, 6))
top_5_districts.plot(kind='bar', color='skyblue')
mtplot.xlabel('District')
mtplot.ylabel('Revenue Growth (%)')
mtplot.title('Top 5 Districts with Highest Document Registration Revenue Growth (FY 2019 to FY 2022)')
mtplot.xticks(rotation=45)
mtplot.tight_layout()

# Show the bar chart
mtplot.show()
```

**Q2:How does the revenue generated from document registration compare
to the revenue generated from e-stamp challans across districts? List
down the top 5 districts where e-stamps revenue contributes
significantly more to the revenue than the documents in FY 2022?**

```python
# Filter data for FY 2022
Data_2022 = Data[Data['month'].between('2022-04-01', '2023-03-31')]

# Group by district and calculate total revenue for documents and e-stamps in FY 2022
total_document_revenue_2022 = Data_2022.groupby('district')['documents_registered_rev'].sum()
total_estamps_revenue_2022 = Data_2022.groupby('district')['estamps_challans_rev'].sum()

# Calculate the ratio of e-stamps revenue to document revenue
revenue_ratio = total_estamps_revenue_2022 / total_document_revenue_2022

# Filter districts where e-stamps revenue contributes significantly more (e.g., ratio > 1)
significant_estamps_contributors = revenue_ratio[revenue_ratio > 1]

# Sort the districts by the ratio in descending order and get the top 5
top_5_districts = significant_estamps_contributors.sort_values(ascending=False).head(5)

# Print the top 5 districts where e-stamps revenue contributes significantly more
print("Top 5 Districts with Significant E-Stamps Revenue Contribution in FY 2022:")
print(top_5_districts)
```

```python
mtplot.figure(figsize=(6, 5))
top_5_districts.plot(kind='bar', color='yellow')
mtplot.xlabel('District')
mtplot.ylabel('E-Stamps Revenue to Document Revenue Ratio')
mtplot.title('Top 5 Districts with Significant E-Stamps Revenue Contribution in FY 2022')
mtplot.xticks(rotation=45)
mtplot.tight_layout()

# Show the bar chart
mtplot.show()
```

**Q3:Is there any alteration of e-Stamp challan count and document
registration count pattern since the implementation of e-Stamp
challan? If so, what suggestions would you propose to the
government?**

```python
Data['month'] = pd.to_datetime(Data['month'])

# Define the e-stamp implementation date as a datetime object
estamp_implementation_date = pd.to_datetime('2022-01-01')
# Filter data for the relevant columns and date range
Data_before_estamp = Data[Data['month'] < estamp_implementation_date]
Data_after_estamp = Data[Data['month'] >= estamp_implementation_date]

# Group by month and calculate the total counts for e-Stamp challans and document registration
estamp_counts_before = Data_before_estamp.groupby('month')['estamps_challans_cnt'].sum()
estamp_counts_after = Data_after_estamp.groupby('month')['estamps_challans_cnt'].sum()
document_counts_before = Data_before_estamp.groupby('month')['documents_registered_cnt'].sum()
document_counts_after = Data_after_estamp.groupby('month')['documents_registered_cnt'].sum()

# Create a time series plot to visualize the trends
mtplot.figure(figsize=(12, 6))
mtplot.plot(estamp_counts_before.index, estamp_counts_before, label='e-Stamp Challans (Before)')
mtplot.plot(estamp_counts_after.index, estamp_counts_after, label='e-Stamp Challans (After)')
mtplot.plot(document_counts_before.index, document_counts_before, label='Document Registration (Before)')
mtplot.plot(document_counts_after.index, document_counts_after, label='Document Registration (After)')

mtplot.xlabel('Month')
mtplot.ylabel('Count')
mtplot.title('Trends in e-Stamp Challan and Document Registration Counts')
mtplot.legend()
mtplot.xticks(rotation=45)
mtplot.tight_layout()

mtplot.show()
```

**Q4:Categorize districts into three segments based on their stamp
registration revenue generation during the fiscal year 2021 to 2022.**

```python
# Filter data for the fiscal year 2021 to 2022
Data_2021_to_2022 = Data[Data['month'].between('2021-04-01', '2022-03-31')]

# Calculate the total stamp registration revenue for each district during FY 2021 to 2022
revenue_by_district = Data_2021_to_2022.groupby('district')['documents_registered_rev'].sum().reset_index()

# Determine the revenue segments (e.g., low, medium, high)
quantiles = revenue_by_district['documents_registered_rev'].quantile([0, 0.33, 0.66, 1])
revenue_by_district['segment'] = pd.cut(revenue_by_district['documents_registered_rev'], bins=quantiles, labels=['low', 'medium', 'high'])
print(revenue_by_district)
```

**Q1:Investigate whether there is any correlation between vehicle sales and
specific months or seasons in different districts. Are there any months
or seasons that consistently show higher or lower sales rate, and if yes,
what could be the driving factors? (Consider Fuel-Type category only**

```python
Data1 = pd.read_excel("Data1.xlsx")
Data1
```

```python
Data1.isnull().sum()
```

```python
import seaborn as sns
```

```python
# Filter the columns related to vehicle sales and fuel type
sales_cols = ['dist_code', 'month', 'fuel_type_petrol', 'fuel_type_diesel', 'fuel_type_electric', 'fuel_type_others']
Data1_sales =   Data1[sales_cols]

# Calculate the total sales for each month
Data1_sales['total_sales'] = Data1_sales[['fuel_type_petrol', 'fuel_type_diesel', 'fuel_type_electric', 'fuel_type_others']].sum(axis=1)

# Convert the 'month' column to datetime if it's not already
Data1_sales['month'] = pd.to_datetime(Data1_sales['month'])

# Extract the month and season (quarter) from the date
Data1_sales['sales_month'] = Data1_sales['month'].dt.month
Data1_sales['season'] = Data1_sales['month'].dt.quarter

# Calculate the correlation matrix
correlation_matrix = Data1_sales[['sales_month', 'season', 'total_sales']].corr()

# Create a heatmap to visualize the correlation
mtplot.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
mtplot.title('Correlation Between Sales, Month, and Season (Fuel Type)')
mtplot.show()
```

**Q2:How does the distribution of vehicles vary by vehicle class (MotorCycle, MotorCar, AutoRickshaw, Agriculture) across different districts? Are there any districts with a predominant preference for a specific vehicle class? Consider FY 2022 for analysis.**

```python
Data1_2022 = Data1[Data1['month'].between('2022-04-01', '2023-03-31')]

# Select columns related to vehicle class and district
vehicle_class_cols = ['district', 'vehicleClass_MotorCycle', 'vehicleClass_MotorCar', 'vehicleClass_AutoRickshaw', 'vehicleClass_Agriculture']
Data1_vehicle_class_2022 = Data1_2022[vehicle_class_cols]

# Group by district and calculate the total count for each vehicle class
vehicle_class_distribution = Data1_vehicle_class_2022.groupby('district').sum()

# Find the predominant vehicle class for each district
predominant_vehicle_class = vehicle_class_distribution.idxmax(axis=1)

# Display the distribution of vehicle classes and predominant vehicle class by district
result_Data1 = pd.DataFrame({'Predominant Vehicle Class': predominant_vehicle_class})
result_Data1 = result_Data1.join(vehicle_class_distribution)
print(result_Data1)
```

**Q3:. List down the top 3 and bottom 3 districts that have shown the highest and lowest vehicle sales growth during FY 2022 compared to FY 2021? (Consider and compare categories: Petrol, Diesel and Electric)**

```python
Data1_2021 = Data1[Data1['month'].between('2021-04-01', '2022-03-31')]
Data1_2022 = Data1[Data1['month'].between('2022-04-01', '2023-03-31')]

# Select columns related to fuel type and district
fuel_type_cols = ['district', 'fuel_type_petrol', 'fuel_type_diesel', 'fuel_type_electric']
Data1_fuel_type_2021 = Data1_2021[fuel_type_cols]
Data1_fuel_type_2022 = Data1_2022[fuel_type_cols]

# Calculate the sales growth for each district and fuel type
sales_growth = ((Data1_fuel_type_2022.groupby('district').sum() - Data1_fuel_type_2021.groupby('district').sum()) / Data1_fuel_type_2021.groupby('district').sum()) * 100

# Define a function to get the top and bottom districts for a given fuel type
def get_top_bottom_districts(fuel_type):
    top_districts = sales_growth[fuel_type].nlargest(3)
    bottom_districts = sales_growth[fuel_type].nsmallest(3)
    return top_districts, bottom_districts

# Display the results for each fuel type
fuel_types = ['fuel_type_petrol', 'fuel_type_diesel', 'fuel_type_electric']

for fuel_type in fuel_types:
    top, bottom = get_top_bottom_districts(fuel_type)

    print(f"\nTop 3 Districts with Highest {fuel_type.capitalize()} Sales Growth:")
    print(top)

    print(f"\nBottom 3 Districts with Lowest {fuel_type.capitalize()} Sales Growth:")
    print(bottom)
```
